# Covariance
Covariance indicates the direction of a linear relationship between two variables.

```python
import numpy as np

# Sample data:
# Variable X: Amount of Fertilizer used (in kg)
fertilizer_used = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
# Variable Y: Crop Yield (in kg)
crop_yield = np.array([50, 55, 62, 70, 78, 85, 92, 105, 110, 120])

# np.cov() takes a 2D array where each row is a variable.
# We stack our two 1D arrays as rows in a 2D array.
data = np.stack((fertilizer_used, crop_yield), axis=0)

# The function calculates sample covariance (dividing by n-1).
covariance_matrix = np.cov(data)

print(f"Fertilizer Used: {fertilizer_used}")
print(f"Crop Yield:      {crop_yield}")
print("\n" + "="*40 + "\n")

print("Covariance Matrix")
print(covariance_matrix)

# Interpretation:
# [[ Var(X),  Cov(X,Y) ],
#  [ Cov(Y,X), Var(Y,Y) ]]

# The off-diagonal values are the covariance.
cov_xy = covariance_matrix[0, 1]
print(f"\nCovariance between Fertilizer and Yield: {cov_xy:.2f}")
print("A positive covariance indicates that as fertilizer use increases, yield tends to increase.")
```

# Correlation
Correlation standardizes the relationship on a scale of -1 to +1, making it easier to interpret the strength of the bond.

```python
import matplotlib.pyplot as plt

# --- POSITIVE CORRELATION ---
# Example: Annual Rainfall (mm) vs. Water Level in a Reservoir (meters)
rainfall = np.array([200, 350, 500, 650, 800, 950, 1100, 1250])
reservoir_level = np.array([10.2, 12.5, 15.1, 18.0, 22.3, 25.8, 29.1, 32.5])

# Calculate correlation matrix
pos_corr_matrix = np.corrcoef(rainfall, reservoir_level)
r_positive = pos_corr_matrix[0, 1]

print("Positive Correlation: Rainfall vs. Reservoir Level")
print(f"Correlation Coefficient (r): {r_positive:.4f}")
print("Value near +1 indicates a very strong positive linear relationship.")


# --- NEGATIVE CORRELATION ---
# Example: Car Speed (km/h) vs. Travel Time for 100km (minutes)
car_speed = np.array([30, 40, 50, 60, 70, 80, 90, 100, 110, 120])
travel_time = np.array([200, 150, 120, 100, 85, 75, 66, 60, 54, 50])

# Calculate correlation matrix
neg_corr_matrix = np.corrcoef(car_speed, travel_time)
r_negative = neg_corr_matrix[0, 1]

print("\nNegative Correlation: Car Speed vs. Travel Time")
print(f"Correlation Coefficient (r): {r_negative:.4f}")
print("Value near -1 indicates that as speed increases, travel time decreases significantly.")
```

#Visualization
Seeing the data points helps distinguish between a simple trend (Covariance) and a tight linear fit (Correlation).

```python
# Plotting the relationships
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Positive Correlation
ax1.scatter(rainfall, reservoir_level, color='teal', s=100, edgecolors='black')
ax1.set_title(f'Positive Correlation (r={r_positive:.2f})\nRainfall vs Reservoir Level')
ax1.set_xlabel('Rainfall (mm)')
ax1.set_ylabel('Reservoir Level (meters)')
ax1.grid(True, linestyle='--', alpha=0.7)

# Plot 2: Negative Correlation
ax2.scatter(car_speed, travel_time, color='orange', s=100, edgecolors='black')
ax2.set_title(f'Negative Correlation (r={r_negative:.2f})\nSpeed vs Travel Time')
ax2.set_xlabel('Car Speed (km/h)')
ax2.set_ylabel('Travel Time (minutes)')
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Key Differences to Remember:
Covariance (64.78 in our first example): Tells us that the relationship is "Positive." However, if we changed fertilizer from kg to grams, this number would explode in size, even though the actual relationship hasn't changed.

Correlation (0.95 or -0.98): This is the "refined" version. It tells us how close the dots are to a straight line. Because it's always between -1 and 1, we can immediately tell that Rainfall is a very reliable predictor of Reservoir Level.
